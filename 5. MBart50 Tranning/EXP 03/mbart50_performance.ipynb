```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Excel file
file_path = 'bleu_scores.xlsx'
df = pd.read_excel(file_path)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(df['Epoch'], df['BLEU Score'], marker='o', color='green', linestyle='-')

plt.title('BLEU Score over Epochs', fontsize=16)
plt.xlabel('Epoch', fontsize=14)
plt.ylabel('BLEU Score', fontsize=14)
plt.grid(True)
plt.tight_layout()

# Save the plot
plt.savefig('bleu_score_plot.png', dpi=300)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# BLEU scores for each split
epochs_split1 = [1,2,3,4, 5]
bleu_split1 = [13.55,14.6,15.28, 15.53,15.66]

epochs_split2 = [1,2,3,4, 5]
bleu_split2 = [13.67,15.29,16.11,16.77,16.89]

epochs_split3 = [1,2, 3]
bleu_split3 = [12.75, 14.39,15.65]

# Plotting the graphs
plt.figure(figsize=(10, 6))

plt.plot(epochs_split1, bleu_split1, marker='o', label='Split 1: 80%-10%-10%')
plt.plot(epochs_split2, bleu_split2, marker='s', label='Split 2: 90%-10% + External Validation')
plt.plot(epochs_split3, bleu_split3, marker='^', label='Split 3: 80%-20% + External Validation')

# Adding titles and labels
plt.title('Comparison of BLEU Scores Across mBART50 Experiments', fontsize=14)
plt.xlabel('Epoch', fontsize=12)
plt.ylabel('BLEU Score', fontsize=12)
plt.grid(True)
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()
```
